```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Churn.csv')
```

```python
df.head(4)
```

```python
# selecting & splitting the required inpendent and target varible columns into X and y
# also hot-code encode the variables into ( 0 & 1)

X= pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis =1))

y =df['Churn'].apply(lambda x: 1 if x== 'Yes' else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
X_train.head(3)
```

```python
y_train.head(3)
```

### Step 2: Import Dependencies

 import from tensorflow 
- Sequential, load model from tensorflow.keras.models.
- Dense from tensorflow.keras.layers.

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

-----------------------------------------------------------------------------

### Step 3: Build & Compile the model

```python
model = Sequential()
model.add(Dense(units= 32, activation ='relu', input_dim = len(X_train.columns) ))
model.add(Dense(units = 64, activation= 'relu'))
model.add(Dense(units = 1, activation='sigmoid'))

```

```python
model.compile(loss='binary_crossentropy' , optimizer= 'sgd', metrics ='accuracy')
```

- Loss Function: binary_crossentropy is suitable for binary classification and penalizes incorrect predictions logarithmically.
- Optimizer: sgd is a basic yet effective optimization algorithm that updates model parameters iteratively.
- Metrics: accuracy provides a clear measure of model performance during training and testing.

This combination ensures that the model is trained effectively for a binary classification task, using a straightforward optimization algorithm and an easily interpretable performance metric.

### Step 4: FIT, PREDICT and EVALUATE

```python
# we have to covert to NumPy arrays and ensure correct data types

X_train=np.array(X_train).astype('float32')

y_train=np.array(y_train).astype('int')


```

Key Points:
- Features (X_train): Should be converted to float32 since neural networks operate on floating-point numbers.
- Labels (y_train): For binary classification, labels should be integers (0 or 1).

```python
model.fit(X_train, y_train, epochs=200, batch_size =32)
```

```python
X_test=np.array(X_test).astype('float32')

y_test=np.array(y_test).astype('int')
```

```python
# predicting

y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# since the above output is in continuous between 0&1 form we'll convert it into binary outcome [0,1]
y_pred = (y_pred > 0.5).astype('int')
```

```python
print(y_pred)
```

```python
accuracy_score(y_test, y_pred)
```
